```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

from sklearn.preprocessing import scale

from scipy.cluster import hierarchy
```

```python
arrests = sm.datasets.get_rdataset('USArrests', package='datasets').data
arrests.head()
```

```python
# Unscaled
plt.figure(figsize=(16, 5))
plt.title('Hierarchical clustering of USArrests (unscaled)')
hierarchy.dendrogram(hierarchy.complete(arrests),
                    labels=arrests.index, color_threshold=75, orientation='top', leaf_font_size=10)
plt.hlines(75, 0, plt.gca().xaxis.get_data_interval()[1], colors='r', linestyles='dashed')
plt.show()
```

```python
# Scaled
plt.figure(figsize=(16, 5))
plt.title('Hierarchical clustering of USArrests (scaled)')
hierarchy.dendrogram(hierarchy.complete(scale(arrests)),
                    labels=arrests.index, color_threshold=3, orientation='top', leaf_font_size=10)
plt.hlines(3, 0, plt.gca().xaxis.get_data_interval()[1], colors='r', linestyles='dashed')
plt.show()
```

Based on my choice of cutpoint for the dendograms, we still get five clusters for both scaled and unscaled data, but the shape of the dendograms is wildly different. 

The data **must** be scaled because the units of variables are different: `UrbanPop` is a percentage while the rest of the variables are measured in "per 100,000".
