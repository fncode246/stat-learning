````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

college = pd.read_csv('data/College.csv', index_col=0)
college.head()
````

````python
with pd.option_context('float_format', '{:.0f}'.format):
  print(college.describe())
````

````python
sns.set(style='ticks')
sns.pairplot(college)
````

Can also use `scatter_matrix` of Pandas like below to produce scatterplot matrix:
```

```

````python
# fig, ax = plt.subplots(figsize=(14, 8))

plt.figure(figsize=(12, 4))
plt.scatter('Outstate', 'Private', data=college)
plt.xlabel('Outstate tuition')
plt.ylabel('Private')
````

````python
# Binning
college['Elite'] = pd.cut(college['Top10perc'], bins=[0, 50, 100], labels=['No', 'Yes'])
````

````python
college['Elite'].value_counts()
````

````python
plt.figure(figsize=(12, 4))
plt.scatter('Outstate', 'Elite', data=college)
plt.xlabel('Outstate tuition')
plt.ylabel('Elite')
````

````python
fig, axes = plt.subplots(figsize=(12, 8), nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = axes.flatten()

ax1.hist('Accept', bins=60, data=college)
ax1.set_title('No. Applicants accepted')

ax2.hist('perc.alumni', bins=15, data=college)
ax2.set_title('Pecent of alumni who donate')

ax3.hist('S.F.Ratio', bins=10, data=college)
ax3.set_title('Student/Faculty ratio')

ax4.hist('Expend', bins=40, data=college)
ax4.set_title('Expenditure per student')

plt.show()
````

````python
plt.figure(figsize=(12, 6))
plt.scatter('Top10perc', 'Grad.Rate', data=college)
plt.xlabel('New students from top 10% of high school class')
plt.ylabel('Graduation rate')
````

Colleges with most students from top 10% of high school class don't necessarily have the highest graduation rate. `Grad.Rate` > 100 is erroneous.
